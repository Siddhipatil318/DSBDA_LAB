```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("IRIS.csv")
```

```python
# Display first few rows
df.head()
```

```python
#1: List Features and Their Types
# Info about dataset
df.info()

# Separate numeric and categorical features
numeric_features = df.select_dtypes(include=['float64']).columns.tolist()
categorical_features = df.select_dtypes(include=['object']).columns.tolist()

print("Numeric Features:", numeric_features)
print("Categorical Features:", categorical_features)
```

```python

#2: Plot Histogram for Each Feature
# Plot histograms
df[numeric_features].hist(bins=15, figsize=(12, 8), edgecolor='black')
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#3 Create Box Plot for Each Feature
plt.figure(figsize=(12, 8))
for i, column in enumerate(numeric_features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df[column])
    plt.title(f'Box Plot of {column}')
plt.tight_layout()
plt.show()
```
