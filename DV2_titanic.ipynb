```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
# Plot box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic)
plt.title('Age Distribution by Sex and Survival')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survived (0 = No, 1 = Yes)')
plt.show()
```

```python
#--> 2. Observations and Inference:
From the box plot, we can draw the following insights:
Median Age:
Both males and females who survived tend to be younger than those who did not survive.
Female survivors have a slightly lower median age compared to female non-survivors.

Age Spread:
Males have a wider age range compared to females.
Non-survivor males include older individuals, while survivor males are more concentrated in the younger age groups.

Children and Young Survivors:
Some very young passengers (children) survived, suggesting that age may have played a role in prioritizing rescue (e.g., "women and children first").

Gender Advantage:
Female passengers had a higher survival rate overall, and the box plot shows that across nearly all age groups, more females survived than males.


```
